```python
import csv
import tkinter as tk
from tkinter import messagebox, simpledialog

def load_data():
    try:
        with open('personnel.csv', 'r', newline='') as file:
            reader = csv.DictReader(file)
            personnel = list(reader)
        return personnel
    except FileNotFoundError:
        return []

def save_data(personnel):
    with open('personnel.csv', 'w', newline='') as file:
        fieldnames = ['name', 'phone', 'position']
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(personnel)

def add_person(personnel):
    name = input("이름을 입력하세요: ")
    phone = input("전화번호를 입력하세요: ")
    position = input("직급을 입력하세요: ")
    new_person = {'name': name, 'phone': phone, 'position': position}
    personnel.append(new_person)
    print("인원이 추가되었습니다.")

def edit_person(personnel):
    name = input("수정할 인원의 이름을 입력하세요: ")
    for person in personnel:
        if person['name'] == name:
            person['phone'] = input("새로운 전화번호를 입력하세요: ")
            person['position'] = input("새로운 직급을 입력하세요: ")
            print("인원 정보가 수정되었습니다.")
            return
    print("일치하는 인원을 찾을 수 없습니다.")

def delete_person(personnel):
    name = input("삭제할 인원의 이름을 입력하세요: ")
    for person in personnel:
        if person['name'] == name:
            personnel.remove(person)
            print("인원 정보가 삭제되었습니다.")
            return
    print("일치하는 인원을 찾을 수 없습니다.")

def display_personnel(personnel):
    for person in personnel:
        print(f"이름: {person['name']}, 전화번호: {person['phone']}, 직급: {person['position']}")

def authenticate(password):
    root = tk.Tk()
    root.withdraw()
    user_password = simpledialog.askstring("비밀번호 입력", "비밀번호를 입력하세요:", show='*')
    return user_password == password

def main():
    password = "mypassword"  # 비밀번호 설정
    authenticated = authenticate(password)

    if not authenticated:
        messagebox.showerror("인증 실패", "비밀번호가 잘못되었습니다. 프로그램을 종료합니다.")
        return

    personnel = load_data()

    while True:
        print("\n인원 관리 프로그램")
        print("1. 인원 추가")
        print("2. 인원 수정")
        print("3. 인원 삭제")
        print("4. 인원 조회")
        print("5. 종료")
        choice = input("원하는 작업 번호를 입력하세요: ")

        if choice == '1':
            add_person(personnel)
        elif choice == '2':
            edit_person(personnel)
        elif choice == '3':
            delete_person(personnel)
        elif choice == '4':
            display_personnel(personnel)
        elif choice == '5':
            save_data(personnel)
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 선택입니다. 다시 입력하세요.")

if __name__ == "__main__":
    main()
```
